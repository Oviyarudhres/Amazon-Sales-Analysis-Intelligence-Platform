```python
#Rename all columns manually for clarity and consistency
amazon_analysis.columns = ['transaction_id', 'order_date', 'customer_id', 'product_id',
       'product_name', 'category', 'subcategory', 'brand',
       'original_price_inr', 'discount_percent', 'discounted_price_inr',
       'quantity', 'subtotal_inr', 'delivery_charges', 'final_amount_inr',
       'customer_city', 'customer_state', 'customer_tier',
       'customer_spending_tier', 'customer_age_group', 'payment_method',
       'delivery_days', 'delivery_type', 'is_prime_member', 'is_festival_sale',
       'festival_name', 'customer_rating', 'return_status', 'order_month',
       'order_year', 'order_quarter', 'product_weight_kg', 'is_prime_eligible',
       'product_rating']
#Clean up column names for presentation:
# Replace underscores (_) with spaces.
# Convert to Title Case (e.g., 'order_date' â†’ 'Order Date')

amazon_analysis.columns = amazon_analysis.columns.str.replace("_", " ").str.title()
amazon_analysis.columns
```

```python
#analyze all categorical (text/object-type) columns in your dataset â€”
#to see what unique values exist in each, how often they occur, and how diverse the data
object_cols = amazon_analysis.select_dtypes(include='object').columns  # Get only text columns

# Loop through each categorical column
for col in object_cols:
    print("=" * 50)
    print(f"ðŸ“Š Column Name: {col}")
    print("-" * 50)
    
    # Show top 10 most common values and their counts
    value_counts = amazon_analysis[col].value_counts().head(10)
    print(value_counts)
    
    # Show number of unique values for extra insight
    print(f"\nUnique values: {amazon_analysis[col].nunique()}")
    print("=" * 50, "\n")
```

```python
# Change 'Order Date' column to date format
amazon_analysis['Order Date'] = pd.to_datetime(
    amazon_analysis['Order Date'], errors='coerce', format='mixed'
)
#Change these columns to numbers
for col in ['Original Price Inr', 'Delivery Days', 'Customer Rating']:
    amazon_analysis[col] = pd.to_numeric(amazon_analysis[col], errors='coerce')
#Change these columns to True/False values
for col in ['Is Festival Sale', 'Is Prime Member', 'Is Prime Eligible']:
    amazon_analysis[col] = pd.to_numeric(amazon_analysis[col], errors='coerce').astype(bool)
#Print message when done
print("âœ… All columns converted successfully!")
```

```python
amazon_analysis[['delivery_charges', 'customer_rating']] = (
    amazon_analysis[['delivery_charges', 'customer_rating']].fillna(method='bfill')
)
#fills missing values in Delivery Charges and Customer Rating columns by taking the next available non-missing value below in the same column.
```

```python
amazon_analysis[['original_price_inr', 'delivery_days', 'customer_rating']] = (
    amazon_analysis[['original_price_inr', 'delivery_days', 'customer_rating']].fillna(method='bfill')
)
#filling missing values in those 3 columns with the next available value from below (backward fill).
```

```python
amazon_analysis[['festival_name', 'customer_age_group', 'customer_rating']] = (
    amazon_analysis[['festival_name', 'customer_age_group', 'customer_rating']].fillna(method='ffill')
)
#It is replaced with the previous rowâ€™s value (the one just above it in the same column)-forward fill
```

```python
amazon_analysis['festival_name'] = amazon_analysis['festival_name'].fillna(method='bfill')
#fills empty â€œFestival Nameâ€ cells using the next non-empty value below in the same column.
```

```python
amazon_analysis['customer_age_group'] = amazon_analysis['customer_age_group'].fillna(method='bfill')
#fills empty cells in Customer Age Group with the next rowâ€™s value, so you donâ€™t have missing values in that colums
```

```python
amazon_analysis['original_price_inr'] = amazon_analysis['original_price_inr'].fillna(method='ffill')
#fills empty cells in Original Price Inr with the last known value above (forward fill), so there are no missing values in that column.
```

```python
null_columns = amazon_analysis.isnull().sum().pipe(lambda x: x[x > 0])
print(null_columns if not null_columns.empty else "No Null Values")
#If any column has missing values, it shows how many. Otherwise, it says â€œNo Null Values.
```

```python
import pandas as pd

data = {
    'Order Date': ['2025-01-01', '2025-01-02', '2025-01-03'],
    'Customer Age Group': ['Adult', 'Teen', None],
    'Customer City': ['Mumbai', 'Bengaluru', 'delhi ncr'],  # column with variations
    'Original Price Inr': [500, 600, 700]
}

amazon_analysis = pd.DataFrame(data)

# Step 1: Clean column names
# -----------------------------
amazon_analysis.columns = (
    amazon_analysis.columns
    .str.strip()            # remove extra spaces
    .str.lower()            # convert to lowercase
    .str.replace(' ', '_')  # replace spaces with underscores
)

# -----------------------------
# Step 2: Define city mapping dictionary
# -----------------------------
city_map = {
    "Mumbai": "Mumbai",
    "mumba": "Mumbai",
    "MUMBAI": "Mumbai",
    "Bombay": "Mumbai",

    "Delhi": "Delhi",
    "delhi": "Delhi",
    "DELHI": "Delhi",
    "New Delhi": "Delhi",
    "Delhi NCR": "Delhi",

    "Bangalore": "Bangalore",
    "Banglore": "Bangalore",
    "Bengalore": "Bangalore",
    "BANGALORE": "Bangalore",
    "Bengaluru": "Bangalore",

    "Chennai": "Chennai",
    "CHENNAI": "Chennai",
    "chenai": "Chennai",
    "Madras": "Chennai",

    "Kolkata": "Kolkata",
    "kolkata": "Kolkata",
    "KOLKATA": "Kolkata",
    "Calcutta": "Kolkata"
}

# -----------------------------
# Step 3: Standardize city names
# -----------------------------
amazon_analysis['customer_city'] = amazon_analysis['customer_city'].replace(city_map)

# -----------------------------
# Step 4: Verify the results
# -----------------------------
print(amazon_analysis)
print("\nCity Counts:")
print(amazon_analysis['customer_city'].value_counts())

#dictionary is a lookup table to convert messy city names into a clean, consistent format for analysis.
```
